```python
#plot graphics will appear in notebook
%matplotlib inline  
#plotting the graphs
from matplotlib import pyplot as plt    
#reading the files
import pandas as pd    
```

```python
#csv file has been read using pandas
Data=pd.read_csv(r"D:\Ramya\candy-power-ranking\candy-power-ranking\candy-data.csv")
```

```python
#Multiple line chart function
def line_chart(data,title,xlabel):
    for each_line in data:
        plt.plot(each_line['xdata'],each_line['ydata'],label=each_line['label'])  #creates a figure,plotting area,decorates plot with labels
    plt.xticks(rotation=90,fontweight="bold",fontsize = '10')  #x-axis names
    plt.xlabel(xlabel,color='black',fontweight="normal",fontsize = '15') #x- axis label
    Data.head(2)
    plt.title(title,color='brown',fontweight="bold",fontsize = '22') #title of figure
    plt.xlim(25, 10)  # x- axis limitations
    plt.legend()   #describing the elements of the graph 
    plt.show()   #showing graph
    
#passing the arguments into variable
data = [{'xdata': Data['competitorname'],'ydata':Data['sugarpercent'],'label':'Sugar Percent'},{ 'xdata':Data['competitorname'],'ydata':Data['pricepercent'],'label':'Price Percent'}]
xlabel='Competitor Name'


#calling line_chart function
line_chart(data,title='Sugar-Price Comparision',xlabel=xlabel)
```

```python
#creating pie_chart funcion
def pie_chart(data,labels,title,myexplode):
    plt.pie(data,labels=labels,explode=myexplode,shadow=True,autopct='%1.2f%%',startangle = 60)   #creates a figure,plotting area,decorates plot with labels
    plt.title(title,color='brown',fontweight="bold",fontsize = '25')    #title of figure
    plt.show()   #showing graph

labels=Data.iloc[:,2:9].columns       #names of the candies
Names=Data.iloc[:,2:9].sum().values #total sum of the each candy 


#calling pie_chart function
pie_chart(Names,labels,'Types of Candy',[0.1,0.1,0.1,0.3,0.3,0.1,0.1])
```

```python
#creating bar_chart function
def bar_chart(xdata,ydata,xlabel,ylabel,tick_label,width,title):
    plt.bar(xdata, ydata, tick_label = tick_label,width=width,color='blue')   #creates a figure,plotting area,decorates plot with labels
    plt.xlabel(xlabel,color='black',fontweight="normal",fontsize = '13')  #x-axis label
    plt.xticks(rotation=90)   #x- axis names and text angle to 90 degrees
    plt.xlim(50, 30)  #limitations of x axis
    plt.ylabel(ylabel,color='black',fontweight="normal",fontsize = '13')  #y-axis label
    plt.title(title,color='Brown',fontweight="bold",fontsize = '25')  #title of figure
    plt.show()  #showing graph

    
#calling bar_chart function  
bar_chart(Data['competitorname'],Data['winpercent'],'competitor name','win percent',Data['competitorname'],0.8,'Win Calculator')    
```
